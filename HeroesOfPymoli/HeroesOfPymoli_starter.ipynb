```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Counting unique players and created data table for output
UniquePlayers = purchase_data["SN"].unique()
Total = len(UniquePlayers)

Total_Players_df = pd.DataFrame({"Total Players" : [Total]})
Total_Players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Analysing unique items, average price, no. purhcases and total revenue
UniqueItems = len(purchase_data["Item ID"].unique())
AveragePrice = purchase_data["Price"].mean()
NoPurchases = len(purchase_data["Price"])
TotalRevenue = purchase_data["Price"].sum()

#Creating data table for output - changing format for $values
PurchasingAnalysis_df = pd.DataFrame({"Number of Unique Items" : [UniqueItems],
                                  "Average Price" : ["${:.2f}".format(AveragePrice)],
                                  "Number of Purchases" : [NoPurchases],
                                  "Total Revenue" : ["${:.2f}".format(TotalRevenue)]})
PurchasingAnalysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Group by gender, creating count and present it correctly
Gender_df = purchase_data.groupby(["Gender"])
GenderCount = Gender_df["SN"].nunique()
#Creating the player percent for gender groups
GPercent = GenderCount/Total*100

#Creating data table for output - changing format for %values, sorting hight to low
GenderDemographic_df = pd.DataFrame({"Total Count": GenderCount,
                                     "Percentage of Players": GPercent})
GenderDemographic_df["Percentage of Players"] = GenderDemographic_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)
GenderDemographic_df = GenderDemographic_df.sort_values("Total Count", ascending=False)
GenderDemographic_df = GenderDemographic_df.rename_axis([""])
GenderDemographic_df

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Creating variables for analysis
GenderAnalysis_df = purchase_data.groupby("Gender")
purchasecount = GenderAnalysis_df["Purchase ID"].count()
averageprice = GenderAnalysis_df["Price"].mean()
totalprice = GenderAnalysis_df["Price"].sum()
avgtotal = totalprice/GenderCount

#Creating data table for output - changing format for $values
GenderAnalysis_df = pd.DataFrame({"Purchase Count": purchasecount,
                                  "Average Purchase Price": averageprice,
                                  "Total Purchase Value": totalprice,
                                  "Avg Total Purchase per Person": avgtotal})
GenderAnalysis_df["Average Purchase Price"] = GenderAnalysis_df ["Average Purchase Price"].astype(float).map("${:.2f}".format)
GenderAnalysis_df["Total Purchase Value"] = GenderAnalysis_df ["Total Purchase Value"].astype(float).map("${:.2f}".format)
GenderAnalysis_df["Avg Total Purchase per Person"] = GenderAnalysis_df ["Avg Total Purchase per Person"].astype(float).map("${:.2f}".format)
GenderAnalysis_df 
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Creating bins for age groups
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]

# Labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39", "40+"]

#Cut labels and add a column
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels = group_labels, include_lowest=True)

#Group by age, creating count and present it correctly
age_group_df = purchase_data.groupby("Age Ranges")
agecount = age_group_df["SN"].nunique()

#Creating the player percent for age groups
agepercent= agecount/Total*100

#Creating data table for output - changing format for %values
AgeDemographic_df = pd.DataFrame({"Total Count": agecount,
                                  "Percentage of Players": agepercent})
AgeDemographic_df["Percentage of Players"] =AgeDemographic_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)
AgeDemographic_df = AgeDemographic_df.rename_axis([""])
AgeDemographic_df


```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Creating variables for analysis
AgeAnalysis_df = purchase_data.groupby("Age Ranges")
purchasecounta = AgeAnalysis_df["Purchase ID"].count()
averagepricea = AgeAnalysis_df["Price"].mean()
totalpricea = AgeAnalysis_df["Price"].sum()
avgtotala = totalpricea/agecount

#Creating data table for output - changing format for $values
AgeAnalysis_df = pd.DataFrame({"Purchase Count": purchasecounta,
                                  "Average Purchase Price": averagepricea,
                                  "Total Purchase Value": totalpricea,
                                  "Avg Total Purchase per Person": avgtotala})
AgeAnalysis_df["Average Purchase Price"] = AgeAnalysis_df["Average Purchase Price"].astype(float).map("${:.2f}".format)
AgeAnalysis_df["Total Purchase Value"] = AgeAnalysis_df["Total Purchase Value"].astype(float).map("${:.2f}".format)
AgeAnalysis_df["Avg Total Purchase per Person"] = AgeAnalysis_df["Avg Total Purchase per Person"].astype(float).map("${:.2f}".format)
AgeAnalysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Creating variables for analysis
Top_df = purchase_data.groupby("SN")
purchasecountt = Top_df["Purchase ID"].count()
averagepricet = Top_df["Price"].mean()
totalpricet = Top_df["Price"].sum()

#Creating data table for output - changing format for $values
TopSpenders_df = pd.DataFrame({"Purchase Count": purchasecountt,
                                  "Average Purchase Price": averagepricet,
                                  "Total Purchase Value": totalpricet})
TopSpenders_df = TopSpenders_df.sort_values("Total Purchase Value", ascending=False)
TopSpenders_df["Average Purchase Price"] = TopSpenders_df["Average Purchase Price"].astype(float).map("${:.2f}".format)
TopSpenders_df["Total Purchase Value"] = TopSpenders_df["Total Purchase Value"].astype(float).map("${:.2f}".format)
TopSpenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Pop_df = purchase_data.groupby(["Item ID","Item Name"])
purchasecounti = Pop_df["Purchase ID"].count()
itemprice = Pop_df["Price"].mean()
totalpricei = Pop_df["Price"].sum()

#Creating data table for output - changing format for $values
PopularItems_df = pd.DataFrame({"Purchase Count": purchasecounti,
                                  "Item Price": itemprice,
                                  "Total Purchase Value": totalpricei})
PopularItems_df = PopularItems_df.sort_values("Purchase Count", ascending=False)
PopularItems_df["Item Price"] = PopularItems_df["Item Price"].astype(float).map("${:.2f}".format)
PopularItems_df["Total Purchase Value"] = PopularItems_df["Total Purchase Value"].astype(float).map("${:.2f}".format)
PopularItems_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Creating data table for output - changing format for $values
ProfitItems_df = pd.DataFrame({"Purchase Count": purchasecounti,
                                  "Item Price": itemprice,
                                  "Total Purchase Value": totalpricei})
ProfitItems_df = ProfitItems_df.sort_values("Total Purchase Value", ascending=False)
ProfitItems_df["Item Price"] = ProfitItems_df["Item Price"].astype(float).map("${:.2f}".format)
ProfitItems_df["Total Purchase Value"] = ProfitItems_df["Total Purchase Value"].astype(float).map("${:.2f}".format)
ProfitItems_df.head()
```
